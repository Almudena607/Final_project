```python
#from dataset import labels
import cv2 as cv
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# loading images-> try to only import the variable
# from dataset.exploratory_analysis import dataset_path
import os
import glob
import numpy as np
dataset_path =  os.path.abspath("dataset")
globpath = f"{dataset_path}/**/*.jpg"
pictures = np.hstack([np.array(glob.glob(globpath, recursive = True))])
pictures[:10]
```

```python
# opening the images in gray scale -> sale en verde nada m√°s(?)
def opening_images(images):
    for im in images:
        img_read = cv.imread(im, 0)
    return img_read
```

```python
opened_images = opening_images(pictures)
```

```python
plt.subplot(1, 1, 1)
plt.imshow(opened_images)
plt.show()
```
